```python
import numpy as np
from scipy.ndimage import convolve

image = np.array([[2, 1, 2, 3, 1],
                  [3, 9, 1, 1, 4],
                  [4, 5, 0, 7, 0]])

kernel = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

convolved_image_s = convolve(image, kernel, mode='constant')
convolved_image_s
```

### is the negative of what we got when we did it by hand!

```python
import matplotlib.pyplot as plt

def plot_images(before, after, before_title="Before Convolution", after_title="After Convolution"):
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    
    axs[0].imshow(before, cmap='gray', interpolation='nearest')
    axs[0].set_title(before_title)
    axs[0].set_xticks([])
    axs[0].set_yticks([])
    
    axs[1].imshow(after, cmap='gray', interpolation='nearest')
    axs[1].set_title(after_title)
    axs[1].set_xticks([])
    axs[1].set_yticks([])
    
    plt.show()

plot_images(image, convolved_image_s)
```

### The convolve image seem to capture the vertical edges of the image, and ignoring the horizontal edges. This is what we expect from sobels kernel, as it is a vertical edge detector.

```python
import cv2

# Convert the image to a format suitable for OpenCV (needs float32 for custom kernels)
image_float32 = np.float32(image)

# Define the kernel (also as float32)
kernel_float32 = np.float32(kernel)

# Apply the convolution using OpenCV's filter2D function
convolved_image_cv = cv2.filter2D(image_float32, -1, kernel_float32, borderType=cv2.BORDER_CONSTANT)

# Convert back to regular integer format for display, if necessary
convolved_image_cv_int = np.int32(convolved_image_cv)
convolved_image_cv_int
```

#### OpenCV seem to give the same result as we got when we calulated by hand. 

```python
plot_images(image, convolved_image_cv_int, before_title="Original Image", after_title="Convolved Image (OpenCV)")
```

### Again, the convolve image seem to capture the horizontal edges of the image, and ignoring the vertical edges. But the image is a negative version of the one from scipy. As we can see in the next cell.

```python
plot_images(image, -convolved_image_s, before_title="Original Image", after_title="Negative Of Convolved Image (Scipy)")
```
